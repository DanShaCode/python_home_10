```python
# Создать новый столбец в таблице с
# пингвинами, который будет отвечать за
# показатель длины клюва пингвина.

# high - длинный(от 42)
# middle - средний(от 35 до 42)
# low - маленький(до 35)
```

```python
import pandas as pd
import seaborn as sns
```

```python
pn = sns.load_dataset("penguins")
pn.head()
```

```python
pn.columns
```

```python
pn.loc[pn['bill_length_mm'] < 35, 'bill_size'] = 'Small'
pn.loc[(pn['bill_length_mm'] > 34) & (pn['bill_length_mm'] < 42), 'bill_size'] = 'Medium'
pn.loc[pn['bill_length_mm'] > 41, 'bill_size'] = 'Big'
```

```python
pn.columns
```

```python
pn[pn['bill_size'] == 'Small']
```

```python
pn[pn['bill_size'] == 'Medium']
```

```python
pn[pn['bill_size'] == 'Big']
```
